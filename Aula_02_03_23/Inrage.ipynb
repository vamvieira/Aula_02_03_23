```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('bolinha.png')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Definio dos valores minimos e maximos da mascara
# o magenta tem h=300 mais ou menos, ou 150 para o opencv
image_lower_hsv = np.array([130, 100, 0])
image_upper_hsv = np.array([175, 255, 255])

mask_hsv = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)

plt.subplot()
plt.imshow(img_rgb)
plt.show()
plt.subplot()
plt.imshow(mask_hsv, cmap="Greys_r", vmin=0, vmax=255)
plt.show()
```

```python
img = cv2.imread('melancia.png')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

image_lower_hsv = np.array([0, 100, 200])
image_upper_hsv = np.array([255, 255, 255])

mask_hsv = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)

plt.subplot()
plt.imshow(img_rgb)
plt.show()
plt.subplot()
plt.imshow(mask_hsv, cmap="Greys_r", vmin=0, vmax=255)
plt.show()
```
